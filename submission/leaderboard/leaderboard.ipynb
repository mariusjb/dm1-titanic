```python
import pandas as pd
from scipy.stats import gaussian_kde, percentileofscore
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
best_public_score = 0.78947
leaderboard = pd.read_csv(f"public_leaderboard.csv", usecols=['TeamId', 'Score'], index_col=0)
```

```python
score = leaderboard.Score
kde = gaussian_kde(score)
xs = np.linspace(score.min(), score.max(), 1000)
pdf = kde.evaluate(xs)
```

```python
plt.figure()
plt.hist(score, bins=100, density=1.0)
perc_of_score = round(percentileofscore(score, best_public_score), 2)
plt.axvline(x = best_public_score, color = 'orange', label = f'score: {best_public_score} [percentile: {perc_of_score}]', linewidth=1.0, linestyle='--')
plt.plot(xs, pdf, label="kde estimated PDF", \
 color="r")
plt.axis([0.6, 0.9, 0, 50])
plt.legend()
plt.show()

print()
```
